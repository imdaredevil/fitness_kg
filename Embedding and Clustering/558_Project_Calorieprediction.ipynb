```python
def score_model(ranks):
  scores = {}
  mrr = mrr_score(ranks)
  print("MRR: %.2f" % (mrr))

  hits_10 = hits_at_n_score(ranks, n=10)
  print("Hits@10: %.2f" % (hits_10))
  hits_3 = hits_at_n_score(ranks, n=3)
  print("Hits@3: %.2f" % (hits_3))
  hits_1 = hits_at_n_score(ranks, n=1)
  print("Hits@1: %.2f" % (hits_1))
```

```python
param_dict = {"batches_count":100, 
                "seed":0, 
                "epochs":200, 
                "k":150, 
                "eta":5,
                "optimizer":'adam', 
                "optimizer_params":{'lr':1e-3},
                "loss":'multiclass_nll', 
                "regularizer":'LP', 
                "regularizer_params":{'p':3, 'lambda':1e-5}, 
                "verbose":True}
```

```python
def evaluate_model(modelName, test_dataset):
  model_new = modelName(**param_dict)
  print(model_new)
  model_new.fit(X_train, early_stopping = False)
  ranks = evaluate_performance(test_dataset, 
                             model=model_new, 
                             filter_triples=positives_filter,   # Corruption strategy filter defined above 
                             corrupt_side='s+o', # corrupt subj and obj separately while evaluating
                             verbose=True)
  score_model(ranks)

  return model_new
```

# 4. Clustering

```python
from statistics import mean, stdev
from itertools import combinations
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
from ampligraph.discovery import find_clusters
```

```python
def perform_agglomerative_clus(distance_threshold_val):
  cluster_algo = AgglomerativeClustering(n_clusters=None, linkage='ward', distance_threshold= distance_threshold_val)
  clusters = find_clusters(entities, model, cluster_algo, mode='entity')

  cluster_lookup = {}
  for cluster, entity in zip(clusters, entities):
    cluster_lookup[entity] = cluster
  return cluster_lookup
```

```python
def evaluate_clusters_New():
  
  evaluation_metric = {"DistanceThreshold": [], "NumClusters": [],"Calories_std_Intra": [], "Met_std_Intra": [], "Calories_std_Inter": [], "Met_std_Inter": []}

  for dis_threshold in range(25, 0, -2):
    # print("processing: ", dis_threshold)

    cluster_lookup = perform_agglomerative_clus(dis_threshold)
    clustered_entities = pd.DataFrame.from_dict(cluster_lookup, orient="index").reset_index().rename(columns={0:'Cluster', 'index':'entity'})
    clustered_with_calories = clustered_entities.merge(cardio[["name"	,	"intensity",	"calories",	"met"]], left_on="entity", right_on="name" , how="left").dropna()

    num_clusters = len(clustered_entities.Cluster.unique())
    calories_std_list = []
    met_std_list = []

    for cluster in clustered_with_calories.Cluster.unique():
      filtered = clustered_with_calories[clustered_with_calories.Cluster==cluster]

      cal_diff_intra = []
      met_diff_intra = []

      if len(filtered.calories.values)>1:
        for cal1, cal2 in combinations(filtered.calories.values, 2):
          cal_diff_intra.append(abs(cal1-cal2))
        for met1, met2 in combinations(filtered.met.values, 2):
          met_diff_intra.append(abs(met1-met2))
      else:
        cal_diff_intra.append(0)
        met_diff_intra.append(0)

      calories_std_list.append(mean(cal_diff_intra))
      met_std_list.append(mean(met_diff_intra))

    calorie_inter_cluster = []
    for calory1, colory2 in combinations(calories_std_list,2):
      calorie_inter_cluster.append(abs(calory1-colory2))

    met_inter_cluster = []
    for met1, met2 in combinations(met_std_list,2):
      met_inter_cluster.append(abs(met1-met2))

      # print("Cluster: {} => Calories_std: {}, Met_std: {}".format(cluster, calories_std, met_std))
    evaluation_metric["DistanceThreshold"].append(dis_threshold)
    evaluation_metric["NumClusters"].append(num_clusters)
    evaluation_metric["Calories_std_Intra"].append(mean(calories_std_list) if len(calories_std_list)>0 else 0)
    evaluation_metric["Met_std_Intra"].append(mean(met_std_list) if len(met_std_list)>0 else 0)
    evaluation_metric["Calories_std_Inter"].append(mean(calorie_inter_cluster) if len(calorie_inter_cluster)>0 else 0)
    evaluation_metric["Met_std_Inter"].append(mean(met_inter_cluster) if len(met_inter_cluster)>0 else 0)

  evaluation_metric_df = pd.DataFrame.from_dict(evaluation_metric)
  return evaluation_metric_df
```

```python
#
#ComplEx, DistMult, TransE, HolE, ConvE, ConvKB
model = ComplEx(batches_count=50,
                epochs=300,
                k=100,
                eta=20,
                optimizer='adam',
                optimizer_params={'lr':1e-4},
                loss='multiclass_nll',
                regularizer='LP',
                regularizer_params={'p':3, 'lambda':1e-5},
                seed=0,
                verbose=True)
model.fit(X)
#complex 1.7
```

```python
workout_embeddings = model.get_embeddings(entities, embedding_type='entity')
```

```python
embeddings_2d = PCA(n_components=2).fit_transform(np.array([i for i in workout_embeddings]))
```
