# Information Extraction

```python
yoga_pose_dict = {
                 'english_Name':[], 'sanskrit_Name':[], 'pose_Type':[],'url': []}

with open('..\Data_Scrape_Yoga\yoga\Extracted\poseTable.jsonl', 'r', encoding="utf8") as json_file:
    json_list = list(json_file)
    
for json_str in json_list:
    result = json.loads(json_str)
    
    yoga_pose_dict['english_Name'].append(result['english_Name'])
    yoga_pose_dict['sanskrit_Name'].append(result['sanskrit_Name'])
    yoga_pose_dict['pose_Type'].append(result['pose_Type'])
    yoga_pose_dict['url'].append(result['url'])
```

```python
yoga_pose_df = pd.DataFrame.from_dict(yoga_pose_dict)
```

```python
yoga_pose_df.head()
```

```python
def find_positions(list_val):
    position_list = []
    for position in list_val:
        if 'Seated' in position:
            position_list.append('sit')
        elif 'Standing' in position:
            position_list.append('stand')
    return position_list
```

```python
yoga_pose_df['position'] = yoga_pose_df.pose_Type.apply(lambda x:find_positions(x))
```
